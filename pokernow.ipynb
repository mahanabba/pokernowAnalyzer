```python
def pokerstats(fpath):
  df = pd.read_csv(fpath)
  s = df["entry"]
  pots = s[s.str.contains(r'collected')==True]
  pattern = "(.*?) collected (.*?) from"
  pattern2 = "(.*?) collected (.*?) from pot with (.*)"
  allwins = []
  for pot in pots:
    try:
      allwins.append([re.search(pattern2, pot).group(3), re.search(pattern2, pot).group(2), re.search(pattern2, pot).group(1).split()[0].strip('"')])
    except:
      allwins.append(["Won before Showdown", re.search(pattern, pot).group(2), re.search(pattern, pot).group(1).split()[0].strip('"')])
  allwins.sort(key=lambda x: float(x[1]), reverse=True)
  allwinsdf = pd.DataFrame(allwins, columns=["Hand", "Size", "Winner"])
  statdic = {}
  for winner in allwinsdf["Winner"].unique():
    statdic[winner] = {}
    statdic[winner]["percentnoshowdown"] = str(round((len(allwinsdf.loc[(allwinsdf["Winner"] == winner) & (allwinsdf["Hand"] == "Won before Showdown")]) / len(allwinsdf.loc[allwinsdf["Winner"] == winner])) * 100, 2)) + "%"
    statdic[winner]["moneywithshowdown"] = "$" + str(round(allwinsdf.loc[(allwinsdf["Winner"] == winner) & (allwinsdf["Hand"] != "Won before Showdown")]["Size"].astype(float).sum(), 2))
    statdic[winner]["moneynoshowdown"] = "$" + str(round(allwinsdf.loc[(allwinsdf["Winner"] == winner) & (allwinsdf["Hand"] == "Won before Showdown")]["Size"].astype(float).sum(), 2))
  
  return allwinsdf, statdic
 
```
